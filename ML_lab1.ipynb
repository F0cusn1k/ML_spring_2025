```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
df = pd.DataFrame(data= np.c_[housing['data'], housing['target']],
                     columns= housing['feature_names'] + ['target'])
df
```

```python
df = df.rename(columns={'target':'MedHouseVal'})
```

```python
df.describe().T
```

```python
df.info()
```

```python
df.hist(bins=50, figsize=(20, 15))
```

```python
temp_df = df.drop('Population', axis=1)
plt.figure(figsize=(20, 10))
sns.boxplot(data=temp_df)
plt.xticks(rotation=45, ha="right")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['MedHouseVal'], kde=True, bins=50)
plt.show()
```

```python
sns.boxplot(y=df['MedHouseVal'])
```

```python
cm = df.corr()
print(cm)

plt.figure(figsize=(12, 10))
sns.heatmap(cm, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='MedInc', y='MedHouseVal', data=df, alpha=0.5)
plt.show()
```
